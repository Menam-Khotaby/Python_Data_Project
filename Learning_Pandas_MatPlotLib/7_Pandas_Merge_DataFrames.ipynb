```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df["job_posted_date"])
```

```python
df_us= df[df['job_country'] == 'United States'].copy()
df_us["job_posted_month"] = df_us["job_posted_date"].dt.strftime("%B")
df_us_pivot = df_us.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
df_us_pivot.reset_index(inplace=True)
```

```python
df_us_pivot["month no"] = pd.to_datetime(df_us_pivot["job_posted_month"] , format="%B").dt.month
df_us_pivot.sort_values(by=["month no"], inplace=True)
df_us_pivot.set_index("job_posted_month", inplace=True)
df_us_pivot.drop(columns="month no", inplace=True)
df_us_pivot
```

```python
df_programming = pd.read_csv("https://lukeb.co/software_csv").set_index("job_posted_month")
```

```python
df_us_merged = df_us_pivot.merge(df_programming, on="job_posted_month")
```

```python
top_5 = (
    df_us_merged
         .sum()
         .sort_values(ascending=False)
         .head(5)
         .index
         .tolist()
         )
top_5
```

```python
df_us_merged[top_5].plot(kind="line")
plt.title("Monthly Job Postings for Top 5 Tech Roles")
plt.xlabel("2023")
plt.ylabel("Jobs count")
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
